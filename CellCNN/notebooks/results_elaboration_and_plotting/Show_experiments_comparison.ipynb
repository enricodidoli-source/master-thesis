```python
decache_files = ['show_results']

from utils import remove_from_cache
remove_from_cache(decache_files)

from show_results import show_metrics_mean_std, retrieve_mean_std
```

```python
plot_exp = '5_plots'

thesis_images_dir = f'{cellcnn_path}/experiments/experiment_{plot_exp}/thesis_images'
os.makedirs(thesis_images_dir, exist_ok=True)

type_classification = ['DIR', 'ROC', 'RES']
exp_names = ['Trial_4_AS_single_split', 'Trial_4_NO_AS_single_split', 'Trial_4_AS_Ensemble', 'Trial_4_NO_AS_Ensemble']
```

```python


tot_exp_std_list = retrieve_mean_std(thesis_images_dir, exp_names, metric = 'rec', function = 'mean')

bar_comp_dir = f'{thesis_images_dir}/bar_comparison/'
os.makedirs(bar_comp_dir, exist_ok = True)

metric_names = ['F1', 'Recall', 'Precision', 'Accuracy']

for m, metric in enumerate(['f1', 'rec', 'pre', 'acc' ]):

    f1_tot_exp_mean_list = retrieve_mean_std(thesis_images_dir, exp_names, metric = metric, function = 'mean')
    f1_tot_exp_std_list = retrieve_mean_std(thesis_images_dir, exp_names, metric = metric, function = 'std')

    name_fig  = f'{metric_names[m]}_bar_comparison'

    show_metrics_mean_std(f1_tot_exp_mean_list, f1_tot_exp_std_list,
                          title = f'{metric_names[m]} Score across Experiments',
                          y_label= f'{metric_names[m]} Score',
                          #save_dir=bar_comp_dir,
                          name_fig = name_fig)
```
