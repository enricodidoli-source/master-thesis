```python
def show_models(i, save_path):
    print(f'Model {i}')
    print('New Generated test datasets')
    with open(f'{save_path}14_10_new_predictions_list_{i}.json', "r", encoding="utf-8") as f:
        imported_new_predictions_1 = json.load(f)

    with open(f'{save_path}14_10_new_results_list_{i}.json', "r", encoding="utf-8") as f:
        imported_new_results_1 = json.load(f)
        
    with open(f'{save_path}14_10_new_test_y_{i}.json', "r", encoding="utf-8") as f:
        imported_new_test_y_1 = json.load(f)
    
    with open(f'{save_path}14_10_original_predictions_list_{i}.json', "r", encoding="utf-8") as f:
        imported_original_predictions_1 = json.load(f)
    
    with open(f'{save_path}14_10_original_results_list_{i}.json', "r", encoding="utf-8") as f:
        imported_original_results_1 = json.load(f)
        
    with open(f'{save_path}14_10_original_test_y_{i}.json', "r", encoding="utf-8") as f:
        imported_original_test_y_1 = json.load(f)


    # show hyperparameters new
    new_best_per_trial_1 = show_hyper(imported_new_results_1, best_3 = False)
    print(new_best_per_trial_1)

    # show predictions new
    new_pred_phenotype_df_1, new_accuracy_list_1 = elaborate_predictions(imported_new_predictions_1, imported_new_test_y_1, results = True)

    index_list = []
    for i, row in enumerate(new_pred_phenotype_df_1.values):
            if 0 in row and 1 in row: 
                if i != len(new_pred_phenotype_df_1.values) - 1:
                    print(f'{i}: {row}, accuracy: {new_accuracy_list_1[i]}')
                    index_list.append(i)
                else:
                    print(f'T: {row}')
    for j in index_list:
        print(f'\nTrial {j} hyperparmeters')
        print(new_best_per_trial_1.iloc[j])
            
    print('Original test datasets')
    # show hyperparameters original
    original_best_per_trial_1 = show_hyper(imported_original_results_1, best_3 = False)
    print(original_new_best_per_trial_1)

    # show predictions original
    original_pred_phenotype_df_1, original_accuracy_list_1 = elaborate_predictions(imported_original_predictions_1, imported_original_test_y_1, results = False)
    
    original_index_list = []
    for i, row in enumerate(original_pred_phenotype_df_1.values):
            if 0 in row and 1 in row: 
                if i != len(original_pred_phenotype_df_1.values) - 1:
                    print(f'{i}: {row}, accuracy: {original_accuracy_list_1[i]}')
                    original_index_list.append(i)
                else:
                    print(f'T: {row}')
                    
    for j in original_index_list:
        print(f'\nTrial {j} hyperparmeters')
        print(original_best_per_trial_1.iloc[j])
        
    
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve
def plot_roc_curve(true_y, y_prob):
    """
    plots the roc curve based of the probabilities
    """

    fpr, tpr, thresholds = roc_curve(true_y, y_prob)
    plt.plot(fpr, tpr)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')

def show_conf_matrix(true_y, y_prob):
    cf_mat = confusion_matrix(true_y, y_prob)
    print('Confusion matrix')
    print(cf_mat)
'''
for trial in pred_phenotype_df.values:
    print(trial)
    print(np.array(new_test_y))
    print(test_blast_perc)
    cf_mat = confusion_matrix(new_test_y, trial)
    print('Confusion matrix')
    print(cf_mat)
    print('')
    plot_roc_curve(new_test_y, trial)
    plt.show()
'''
```
